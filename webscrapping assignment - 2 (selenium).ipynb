Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
#operating the homepage of naukari
driver.get('https://www.naukri.com/')
```

```python
#Automate search
#locating designation search bar
designation_search=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation_search.send_keys("Data Analyst")
```

```python
#finding location webelement for search bar using absolutepath
location_search=driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
location_search.send_keys("Bangalore")
```

```python
#clicking using absolute xpath 

search_btn=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search_btn.click()
```

# Extracting job titles

```python
#scraping job titles
job_titles=[]
title_tags=driver.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
for i in title_tags:
    job_titles.append(i.text)
len(job_titles)
```

```python
#filtering 10 results
job_titles=job_titles[0:10]
job_titles
```

# Extracting job location

```python
#scraping job location
job_location=[]
location_tags=driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft fs12 lh16 locWdth']")
for i in location_tags:
    job_location.append(i.text)
len(job_location)
```

```python
#filtering 10 results
job_location=job_location[0:10]
job_location
```

# Extracting company_name

```python
#scraping company name
company_name=[]
company_tags=driver.find_elements(By.XPATH,"//div[@class='mt-7 companyInfo subheading lh16']")

for i in company_tags:
    company_name.append(i.text.split("\n")[0] )
len(company_name)
```

```python
#filtering 10 results
company_name=company_name[0:10]
company_name
```

# Extracting experience

```python
#scraping experience

experience=[]
experience_tags=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')

for i in experience_tags:
    experience.append(i.text)
len(experience)
```

```python
#filtering 10 results
experience=experience[0:10]
experience
```

```python
#checking length
print(len(job_titles),len(job_location),len(company_name),len(experience))
```

```python
#creating datasets
jobs_naukari=pd.DataFrame({
                            "job_titles":job_titles,
                           "job_location":job_location,
                           "company_name":company_name,
                           "experience":experience
                          })
```

```python
#printing dataset
jobs_naukari
```
