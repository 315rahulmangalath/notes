```python
sal=[15,22,24,25,30,23,24,43,19,20,52,21]
sal.sort()
sal
```

```python
ser=pd.Series(sal)
```

```python
q1=ser.quantile(.25)
```

```python
q3=ser.quantile(.75)
```

```python
#to find the IQR(intermediate quantile range) to find the outliers
iqr=q3-q1
iqr
```

```python
#higer side
higer_side=q3+(1.5*iqr)
lower_side=q3-(1.5*iqr)
```

```python
higer_side
```

```python
lower_side
```

# numerical Data plotting

```python
#loading dataset tips from seaborn library
tips=sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
tips['smoker'].value_counts()
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips,size='size')
plt.show()
```

```python
#categorical data plotting
```

```python
#catplot
sns.catplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#catplot in combination with swarmplotplot
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
plt.show()
```

```python
#stripplot
sns.stripplot(x='size',y='total_bill',data=tips)
plt.show()
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips)
plt.show()
```

```python
#distribution plot
x=randn(1000)
sns.distplot(x)
plt.show()
```

```python
sns.kdeplot(x)
```

```python
#jointplot
x=tips['total_bill']
y=tips['tip']
```

```python
sns.jointplot(x=x,y=y)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```
