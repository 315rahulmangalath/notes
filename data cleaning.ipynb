```python
raw_data={'firstname':['jason',np.nan,'tina','jake','amy','madhu'],
         'lastname':['miller',np.nan,'ali','miler','cooze','patil'],
         'age':[42,np.nan,36,24,73,25],
         'sex':['m',np.nan,'f','m','f','f'],
         'unit1':[4,np.nan,np.nan,2,3,4],
        'unit2' :[3,np.nan,np.nan,4,4,4]}
```

```python
df=pd.DataFrame(raw_data)
```

```python
df
```

```python
#to remove the col with null values
df.dropna(axis=1)
```

```python
#to remove the col with null values
df.dropna(axis=0)
```

```python
df_1=df.dropna()
```

```python
#to remove only rows and col containing null values 

df2=df.dropna(how='all')
df2
```

```python
df_1
```

```python
df["finalscore"]=np.nan
df
```

```python
#to remove only rows and col containing null values (how=all)
df.dropna(how='all',axis=1)
```

```python
df["unit1"].fillna(df["unit1"].mean())
df
```

```python
#fill in missing values with mean of each gender category
df["unit2"].fillna(df.groupby("sex")["unit2"].transform("mean"),inplace=True)
df
```

```python
df[df['age'].notnull() & df['sex'].notnull()]
```

```python
#backfill
df.fillna(method='bfill')
```

```python
#backfill
df.fillna(method='ffill')
```

```python
#to fill values in null values loc
df.fillna(df['unit1'].sum())
```

```python
df['finalscore'].fillna(df["unit1"]+df["unit2"])
df
```
