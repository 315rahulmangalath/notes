```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic_train=pd.read_csv("https://raw.githubusercontent.com/training-ml/Files/main/titanic_train.csv")
```

```python
#to get total description in each columns and therir values
titanic_train.describe()
```

```python
titanic_train.columns
```

```python
new_surv=titanic_train['Survived'].unique()

new_surv
```

```python
new_survived=pd.Categorical(titanic_train["Survived"])
new_survived
```

```python
#renaming the categories of zeroes and ones into died and survived
new_survived=new_survived.rename_categories(["died","Survived"])
new_survived.describe
```

```python
titanic_train['Survived']=new_survived
titanic_train.head()
```

```python
titanic_train['Pclass'].unique()
```

```python
#assigning categorical data to a variable from col titanic_train
new_Pclass=pd.Categorical(titanic_train["Pclass"])
```

```python
#using categorical assigned variable to rename the categories
new_Pclass=new_Pclass.rename_categories(['class1','class2','class3'])
new_Pclass.describe()
```

```python
titanic_train.columns
```

```python
titanic_train['pclass']=new_Pclass
titanic_train.head()
```

```python
char_cabin=titanic_train["Cabin"].astype(str)
char_cabin
```

```python
new_cabin=np.array([cabin[0] for cabin in char_cabin])
```

```python
new_cabin
```

```python
new_cabin=pd.Categorical(new_cabin)
```

```python
titanic_train['cabin']=new_cabin
new_cabin.describe()
```

```python
titanic_train.head()
```

```python
#np.where will get us the index pos of matching condt
missing=np.where(titanic_train['Age'].isnull()==True)
```

```python
age_available=np.where(titanic_train['Age'].isnull()==False)
```

```python
titanic_train.loc[age_available]
```
